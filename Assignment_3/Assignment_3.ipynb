```python
import numpy as np
import pandas as pd
data = pd.read_csv("netflixData.csv")
```

```python
data
```

```python
data['Cast'] = data['Cast'].replace(np.nan,'none')
def no_of_cast(cast):
    if cast == 'none':
        return 0
    else:
        n = cast.split(",")
        return len(n)
```

```python
data["no_of_cast"] = data["Cast"].apply(no_of_cast)
```
